```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sqlite3
import matplotlib.pyplot as plt
```

```python
connection = sqlite3.connect("data_db.db")
```

```python
pd.read_sql("""SELECT *
                    FROM concrete_data ;""", connection)
```

```python
pd.read_sql("""SELECT *
   FROM concrete_data
   WHERE strength IS NULL;""", connection)
```

what is the component of best strength ? 

```python
pd.read_sql("""SELECT *
                    FROM concrete_data ORDER BY strength DESC LIMIT 10
                    ;""", connection)
```

 what is the component of lowest strength ? 

```python
pd.read_sql("""SELECT *
                    FROM concrete_data ORDER BY strength ASC LIMIT 10
                    ;""", connection)
```

 what is minmum strength  ? 

```python
pd.read_sql("""SELECT MIN(strength) AS MINIMUM_STRENGTH
                    FROM concrete_data
                    ;""", connection)
```

 what is Maximum strength  ?

```python
pd.read_sql("""SELECT MAX(strength) AS MAXIMUM_STRENGTH
                    FROM concrete_data
                    ;""", connection)
```

what is the mean of  strength ?

```python
pd.read_sql("""SELECT AVG(strength) AS AVERAGE_STRENGTH
                    FROM concrete_data
                    ;""", connection)
```

what is the age samples ? 

```python
pd.read_sql("""SELECT DISTINCT age 
                    FROM concrete_data
                    ;""", connection)
```

what is the minmun age ?

```python
pd.read_sql("""SELECT MIN(age) AS MINIMUM_AGE
                    FROM concrete_data
                    ;""", connection)
```

what is the maximum age ?

```python
pd.read_sql("""SELECT MAX(age) AS MAXIMUM_AGE
                    FROM concrete_data
                    ;""", connection)
```

what is the mean of age ?

```python
pd.read_sql("""SELECT AVG(age) AS AVERAGE_AGE
                    FROM concrete_data
                    ;""", connection)
```

list the strength between 17 MPa and 28 MPa ?

```python
pd.read_sql("""Select * From concrete_data Where strength BETWEEN 17 AND 28
                    ;""", connection)
```

list the the strength that greater than 28 ?

```python
pd.read_sql("""Select * From concrete_data Where strength > 28
                    ;""", connection)
```

what is the mean of strength of each sample of age?

```python
pd.read_sql("""SELECT age, AVG(strength) FROM concrete_data
   GROUP BY age;""", connection)
```
